```python
from pathlib import Path
import pandas as pd


def compute_averages(folder: Path):
    csv_files = sorted(folder.glob("*.csv"))
    if not csv_files:
        raise FileNotFoundError(f"No CSV files found in {folder}")

    frames = []
    processed_files = 0

    for f in csv_files:
        try:
            df = pd.read_csv(f, sep=";")
        except Exception as e:
            continue

        # Ensure required columns exist
        required_cols = ["Daily query count", "Daily unique domain count"]
        if not all(col in df.columns for col in required_cols):
            continue

        # Keep only the columns we need and ensure they are numeric
        sub = df[required_cols].copy()
        for col in required_cols:
            sub[col] = pd.to_numeric(sub[col], errors="coerce")

        frames.append(sub)
        processed_files += 1

    if not frames:
        raise ValueError("No valid data found in any CSV file.")

    all_data = pd.concat(frames, ignore_index=True)

    avg_daily_query = all_data["Daily query count"].mean()
    avg_unique_domains = all_data["Daily unique domain count"].mean()

    print(f"CSV files processed: {processed_files}")
    print(f"Total rows used: {len(all_data)}\n")
    print(f"Average daily query count (overall): {avg_daily_query:.3f}")
    print(f"Average daily unique domain count (overall): {avg_unique_domains:.3f}")
```

```python
compute_averages(Path("dataset/regular_domain_names/january"))
```
